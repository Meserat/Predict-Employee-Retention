# Explanatory data analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# create dataframe from dataset
df=pd.read_csv("hr_employee_churn_data.csv")
```

```python
# view some data
df.head(4)
```

```python
# check column and it's data type
df.info()
```

```python
# check the shape of the data
df.shape
```

```python
# statsical descrption of the data
df.describe()
```

```python
# check data imbalance
sns.countplot(x='left',data=df)
plt.plot()
```

```python
# salary feature
sns.countplot(x='salary',data=df)
plt.plot()
```

```python
sns.catplot(x='left', col='salary', kind='count',data=df)
```

```python
# salary feature
sns.countplot(x='promotion_last_5years',data=df)
plt.plot()
```

```python
sns.catplot(x='left', col='promotion_last_5years', kind='count',data=df)
```

```python
# work accident
sns.countplot(x='Work_accident',data=df)
plt.plot()
```

```python
# feature left split over feature work accident
sns.catplot(x='left', col='Work_accident', kind='count',data=df)
```

```python
#univariate distribution
sns.displot(df['satisfaction_level'])
```

```python
# box plot distribution for data 
sns.boxplot(x='left',y='number_project',data=df)
```

```python
# box plot distribution for number of project
sns.displot(x='left',y='satisfaction_level',data=df)
```

```python
sns.distplot(df['last_evaluation'])
```

```python
# box plot distribution for data 
sns.boxplot(x='left',y='last_evaluation',data=df)
```
